1. Import confirmed, recovered and deaths data set for a given country

```python
confirmed = pd.read_csv('DATA/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
recovered = pd.read_csv('DATA/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
deaths = pd.read_csv('DATA/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
```

2. Pre-process the datasets

```python
def reformat(df, country, col_name):
    
    df = df[df['Country/Region'] == country]
    index = df.iloc[0].name

    df = df.T
    df = df.rename({index: col_name}, axis = 1)
    # get rid of columns that do not represent the number of recovered/confirmed/deaths 
    df = df.iloc[4:,:]
    
    return df

    #df = df[df[col_name] > 0]
```

```python
recovered = reformat(recovered, 'Switzerland', 'Recovered cases')
confirmed = reformat(confirmed, 'Switzerland', 'Confirmed cases')
deaths = reformat(deaths, 'Switzerland', 'Deaths cases')

recovered.tail()
```

3. Create new data frame recovered per day so we know how many people recovered from the epidemic on a given day


**problem** : Here, I see that recovered_daily contains a negative value, therefore recovered is not fully incremental. ??? Change dataset?

```python
recovered_daily = recovered.diff()
recovered_daily.iloc[0,:] = 0

deaths_daily = deaths.diff()
deaths_daily.iloc[0] = 0
```

4. Generate dataset infected representing the number of infected individuals at a given time t (not the new number of infected, number of actively infected individuals)

$$\text{confirmed}(t)=\text{confirmed}(t)-\sum_{i=\text{day}_0}^{t}\text{recovered_daily(i)}-\sum_{i=\text{day}_0}^{t}\text{deaths_daily}(i)$$

```python
infected_ = []
days_ = []
# i: date (time t)
# row: row
acc_deaths = 0
acc_recovered = 0

for i, row in confirmed.iterrows():
    acc_recovered += recovered_daily.loc[i]['Recovered cases']
    acc_deaths += deaths_daily.loc[i]['Deaths cases']
    
    diff = row['Confirmed cases'] - acc_recovered - acc_deaths
    
    infected_.append(diff)
    days_.append(i)
```

Careful, infected represents active number of infected, not new number of infected

```python
infected = pd.DataFrame({'Date': days_,'Infected Cases': infected_})

recovered = recovered.reset_index().rename({'index':'Date'}, axis = 1)
```

### Exponential smoothing

```python
ax=data['Infected Cases'].plot()
ax.set_xlabel("Day")
ax.set_ylabel("Number of actively infected cases on a given day")
plt.show()
```

```python
def exponential_smoothing(series, alpha):
    result = [series[0]] # first value is same as series
    for n in range(1, len(series)):
        result.append(alpha * series[n] + (1 - alpha) * result[n-1])
    return result
```

```python
df= pd.DataFrame({'Original Data': data['Infected Cases'],'Smoothed Data': exponential_smoothing(data['Infected Cases'],0.8)})
df.plot()
```

Here we run three variants of simple exponential smoothing:
1. In fit1 we do not use the auto optimization but instead choose to explicitly provide the model with the 𝛼=0.2 parameter
2. In fit2 we choose an 𝛼=0.6
3. In fit3 we allow statsmodels to automatically find an optimized 𝛼 value for us.

such that

$$\hat{y}_{T+1|T}=\alpha y_T+\alpha(1-\alpha)y_{T-1}+\alpha(1-\alpha)^2y_{T-2}+...$$



```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt

fit1 = SimpleExpSmoothing(data['Infected Cases']).fit(smoothing_level=0.2,optimized=False)
#fcast1 = fit1.forecast(15).rename(r'$\alpha=0.2$')
fit2 = SimpleExpSmoothing(data['Infected Cases']).fit(smoothing_level=0.6,optimized=False)
#fcast2 = fit2.forecast(15).rename(r'$\alpha=0.6$')
fit3 = SimpleExpSmoothing(data['Infected Cases']).fit(smoothing_level=0.9596)
#fcast3 = fit3.forecast(15).rename(r'$\alpha=0.9596$)

ax = data['Infected Cases'].plot(color='black', figsize=(12,8),label='Original Data', legend=True, fontsize='large')

#fcast1.plot(ax=ax, color='blue', legend=True)
fit1.fittedvalues.plot( ax=ax, color='blue',label=r'$\alpha=0.2$', legend=True)

#fcast2.plot( ax=ax, color='red', legend=True)
fit2.fittedvalues.plot(ax=ax, color='red',label=r'$\alpha=0.6$', legend=True)

#fcast3.plot( ax=ax, color='green', legend=True)
fit3.fittedvalues.plot(ax=ax, color='green',label=r'$\alpha=0.9596$', legend=True)

plt.show()
```

```python
fit1 = Holt(data['Infected Cases']).fit(smoothing_level=0.8, smoothing_slope=0.2, optimized=True)
fcast1 = fit1.forecast(5).rename("Holt's linear trend")
#fit2 = Holt(data['Infected Cases'][30:], exponential=True).fit(smoothing_level=0.8, smoothing_slope=0.2, optimized=False)
#fcast2 = fit2.forecast(5).rename("Exponential trend")
fit3 = Holt(data['Infected Cases'], damped=True).fit(smoothing_level=0.8, smoothing_slope=0.2, optimized=True)
fcast3 = fit3.forecast(5).rename("Additive damped trend")

ax = data['Infected Cases'].plot(color="black",figsize=(12,8))
fit1.fittedvalues.plot(ax=ax, color='blue')
fcast1.plot(ax=ax, color='blue',legend=True)
#fit2.fittedvalues.plot(ax=ax, color='red')
#fcast2.plot(ax=ax, color='red',legend=True)
fit3.fittedvalues.plot(ax=ax, color='green')
fcast3.plot(ax=ax, color='green',legend=True)

plt.show()
```
