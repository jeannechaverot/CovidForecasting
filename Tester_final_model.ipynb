```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from set_axes import set_ax
from models import DeterministicSIndIDRDndRnd as Model
from models import StochasticSIndIDRDndRnd as StochasticModel
```

The equations of motion of the model are the following:

$$
\dot{S} = -\beta\ \frac{S(I + \delta I_{nd})}{N} \\
\dot{I}_{nd} = \beta\ \frac{S(I + \delta I_{nd})}{N} - \alpha I_{nd} - \xi_{nd}I_{nd} - \gamma_{nd}I_{nd} - \xi_2 I_{nd}\\
\dot{I} = \alpha I_{nd} - \gamma I - \xi I\\
\dot{D} = \xi I +\xi_2 I_{nd}\\
\dot{R} = \gamma I\\
\dot{D}_{nd} = \xi_{nd} I_{nd}\\
\dot{R}_{nd} = \gamma_{nd} I_{nd} \\
N = S + I_{nd} + I + D + R + D_{nd} + R_{nd}
$$

```python
params = {'beta' : 2, 
          'alpha' : 0.1, 
          'gamma' : 0.3, 
          'xi' : 0.2,
          'delta' : 0.5,
          'gamma_nd' : 0.5,
          'xi_nd' : 0.05,
          'xi2' : 0.1,
          'initial_population' : (1000, 1, 0, 0, 0, 0, 0)}
model = Model(**params)
model.evolve(days=200)

xlabel = 'Days'
ylabel = 'Deterministic'
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(211)
ax.plot(model.time, model.S, linewidth=2, label=r'$S$')
ax.plot(model.time, model.I_nd, linewidth=2, label=r'$I_{nd}$')
ax.plot(model.time, model.I, linewidth=2, label=r'$I$')
ax.plot(model.time, model.D, linewidth=2, label=r'$D$')
ax.plot(model.time, model.R, linewidth=2, label=r'$R$')
ax.plot(model.time, model.D_nd, linewidth=2, label=r'$D_{nd}$')
ax.plot(model.time, model.R_nd, linewidth=2, label=r'$R_{nd}$')

set_ax(ax, ylabel=ylabel, legend=True)
plt.show()
```
